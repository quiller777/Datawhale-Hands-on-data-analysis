```python
sample = pd.DataFrame(np.random.randn(3,3),   #生成随机DataFrame
                      index=list('213'),
                      columns=list('BCA'))
sample.head()
```

```python
sample.sort_values('B')   #按照B列默认行升序排列 axis =1对列排序，ascending=False为降序
```

```python
sample.sort_index(axis=1) #列索引升序排列
```

```python
sample.sort_index(axis=1,ascending=False)#列索引降序排列
```

```python
sample.sort_values(['A','C'],ascending=False)#两列进行降序排列
```
